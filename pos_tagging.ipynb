```
import nltk
from nltk.corpus import treebank

train_data = treebank.tagged_sents()[:3000]
test_data = treebank.tagged_sents()[3000:]

f = open("feedback_cs2012_1.txt", "r")
tokens = nltk.word_tokenize(f.read())
```

```
# Attempt to Unigram Tagger, Perceptron Tagger & HMMTagger to annotate the text
unigramtagger = nltk.UnigramTagger(train_data)

print("Tagger Info : ")
print(unigramtagger)

print("UGRAM : ", unigramtagger.tag(tokens))

print("UGRAM EVAL : ", unigramtagger.evaluate(test_data))
```

```
perceptrontagger = nltk.PerceptronTagger(train_data)

print("Tagger Info : ")
print(perceptrontagger)

print("PRCPT : ", perceptrontagger.tag(tokens))

print("PRCPT EVAL : ", perceptrontagger.evaluate(test_data))
```

```
hmmtagger = nltk.HiddenMarkovModelTagger.train(train_data)

print("Tagger Info : ")
print(hmmtagger)

print("HMMTR : ", hmmtagger.tag(tokens))

print("HMMTR EVAL : ", hmmtagger.evaluate(test_data))
```
